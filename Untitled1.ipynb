```python
tc = pd.read_csv("time_course_results.csv")

baseline_scatter = go.Scatter(x=list(range(len(tc))), y=tc['base'],
                             line=go.scatter.Line(width=1),
                             opacity=0.9,
                             yaxis='y1',
                             name='Baseline Model')
count_scatter = go.Scatter(x=list(range(len(tc))), y=tc['count'],
                                 line=go.scatter.Line(width=1),
                                 opacity=0.9,
                                 yaxis='y2',
                                 name='Count Model')

correlation_scatter = go.Scatter(x=list(range(len(tc))), y=tc['corr'],
                             line=go.scatter.Line(width=1),
                             opacity=0.9,
                             yaxis='y3',
                             name='Correlation Model')
queue_scatter = go.Scatter(x=list(range(len(tc))), y=tc['queue'],
                                 line=go.scatter.Line(width=1),
                                 opacity=0.9,
                                 yaxis='y4',
                                 name='Queue Model')



layout = go.Layout(height=700, width=1000, titlefont=dict(size=30),
                   font=dict(size=10),
                   title="Models",
                   xaxis=dict(title='Device Type',
                              rangeslider=dict(visible=True),
                              domain=[0.1, 0.9]),
                   yaxis=dict(title='Time',
                               rangemode='tozero',
                             ),
                   yaxis2=dict(title='Time',
                               overlaying='y',
                               rangemode='tozero'),
                   yaxis3=dict(title='Time',
                               overlaying='y',
                               rangemode='tozero'),
                   yaxis4=dict(title='Time',
                               overlaying='y',
                               rangemode='tozero')
                  )

fig = go.Figure(data=[baseline_scatter, count_scatter, correlation_scatter, queue_scatter], layout=layout)

fig.show()
```
